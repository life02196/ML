```python
# Load libraries
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from scipy.stats import mode
import numpy as np
import matplotlib.pyplot as plt

# Load data
faces = datasets.fetch_olivetti_faces()
features = faces.data
labels = faces.target

# Split data into training and test sets
features_train, features_test, labels_train, labels_test = train_test_split(
    features, labels, test_size=0.3, random_state=0)

# Standardize features
scaler = StandardScaler()
features_train_std = scaler.fit_transform(features_train)
features_test_std = scaler.transform(features_test)

# Create k-means object
cluster = KMeans(n_clusters=40, random_state=0, n_init="auto")

# Train model
model = cluster.fit(features_train_std)

# Predict clusters for training and test sets
train_clusters = model.predict(features_train_std)
test_clusters = model.predict(features_test_std)

# Match clusters to true labels
def match_labels(true_labels, cluster_labels):
    labels = np.zeros_like(cluster_labels)
    for i in range(40):
        mask = (cluster_labels == i)
        labels[mask] = mode(true_labels[mask])[0]
    return labels

# Map clusters to labels
train_labels = match_labels(labels_train, train_clusters)
test_labels = match_labels(labels_test, test_clusters)

# Calculate accuracy
train_accuracy = accuracy_score(labels_train, train_labels)
test_accuracy = accuracy_score(labels_test, test_labels)

print(f'Training accuracy: {train_accuracy:.2f}')
print(f'Test accuracy: {test_accuracy:.2f}')

# Visualize the cluster centers
fig, axes = plt.subplots(4, 10, figsize=(15, 8),
                         subplot_kw={'xticks':(), 'yticks':()})
for i, ax in enumerate(axes.flat):
    ax.imshow(model.cluster_centers_[i].reshape(64, 64), cmap='gray')
    ax.set_title(f'Cluster {i+1}')
plt.show()
```

```python
# Load libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MeanShift
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from joblib import Parallel, delayed

# Load the 20 newsgroups dataset
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
data = newsgroups.data
labels = newsgroups.target

# Convert the text data to TF-IDF feature vectors
vectorizer = TfidfVectorizer(max_df=0.5, min_df=5, max_features=5000, stop_words='english')
features = vectorizer.fit_transform(data)

# Perform PCA to reduce dimensionality
pca = PCA(n_components=50)
features_reduced = pca.fit_transform(features.toarray())

# Function to evaluate clustering results
def evaluate_clustering(labels_true, labels_pred):
    homogeneity = homogeneity_score(labels_true, labels_pred)
    completeness = completeness_score(labels_true, labels_pred)
    v_measure = v_measure_score(labels_true, labels_pred)
    return homogeneity, completeness, v_measure

# Create Mean Shift object with a specific bandwidth
bandwidth = 1
cluster = MeanShift(bandwidth=bandwidth, n_jobs=-1)

# Train model
model = cluster.fit(features_reduced)

# Predict clusters
clusters = model.labels_

# Evaluate clustering results
homogeneity, completeness, v_measure = evaluate_clustering(labels, clusters)

# Print the evaluation scores
print(f'MeanShift with bandwidth={bandwidth}, Homogeneity: {homogeneity:.2f}, Completeness: {completeness:.2f}, V-measure: {v_measure:.2f}')

# Plot the results
plt.figure(figsize=(10, 6))
plt.bar(['Homogeneity', 'Completeness', 'V-measure'], [homogeneity, completeness, v_measure], color=['blue', 'orange', 'green'])
plt.ylim(0, 1)
plt.ylabel('Score')
plt.title(f'Clustering Evaluation Scores for MeanShift (bandwidth={bandwidth})')
plt.grid(True)
plt.show()
```

```python
# 載入所需的庫
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score
import numpy as np
import matplotlib.pyplot as plt

# 載入 20 newsgroups 數據集
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))
data = newsgroups.data
labels = newsgroups.target

# 將文本數據轉換為 TF-IDF 特徵向量
vectorizer = TfidfVectorizer(max_df=0.5, min_df=5, stop_words='english')
features = vectorizer.fit_transform(data)

# 定義聚類結果評估函數
def evaluate_clustering(labels_true, labels_pred):
    homogeneity = homogeneity_score(labels_true, labels_pred)
    completeness = completeness_score(labels_true, labels_pred)
    v_measure = v_measure_score(labels_true, labels_pred)
    return homogeneity, completeness, v_measure

# 創建 AgglomerativeClustering 對象
cluster = AgglomerativeClustering(n_clusters=1, linkage='ward')

# 訓練模型
model = cluster.fit(features.toarray())

# 預測聚類
clusters = model.labels_

# 評估聚類結果
homogeneity, completeness, v_measure = evaluate_clustering(labels, clusters)

# 輸出評估分數
print(f'AgglomerativeClustering, Homogeneity: {homogeneity:.2f}, Completeness: {completeness:.2f}, V-measure: {v_measure:.2f}')

# 繪製結果
plt.figure(figsize=(10, 6))
plt.bar(['Homogeneity', 'Completeness', 'V-measure'], [homogeneity, completeness, v_measure], color=['blue', 'orange', 'green'])
plt.ylim(0, 1)
plt.ylabel('Score')
plt.title('Clustering Evaluation Scores for AgglomerativeClustering')
plt.grid(True)
plt.show()
```
